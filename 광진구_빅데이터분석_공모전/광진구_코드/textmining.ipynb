```python
import pandas as pd

import re
```

```python
df = pd.read_csv("./광진구_20230314_14시00분43초.csv")
```

```python
df = df[["title", "content"]]
```

```python
df
```

```python
def text(x):
    x = re.sub(r"[\n\t]", "", x)
    return x
```

```python
df["title"] = df["title"].apply(text)
```

```python
df["content"] = df["content"].apply(text)
```

```python
df
```

```python
df['title'] = df['title'].str.replace('[^가-힣]', ' ', regex = True)
df['title']
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Kkma
from PIL import Image
import numpy as np
```

```python
kkma = Kkma()
```

```python
nouns = df["title"].apply(kkma.nouns)
```

```python
nouns = nouns.explode()
```

```python
df_word = pd.DataFrame({'word' : nouns})
df_word['count'] = df_word['word'].str.len()
df_word = df_word.query('count >= 2')
df_word
```

```python
df_word = df_word.groupby('word', as_index = False).count().sort_values('count', ascending = False)
df_word
```

```python
df_word = df_word.iloc[3:]
```

```python
dic_word = df_word.set_index('word').to_dict()['count']
```

```python
wc = WordCloud(random_state = 123, font_path = 'AppleGothic', width = 400,
               height = 400, background_color = 'white')

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10)) # 크기 지정하기
plt.axis('off') # 축 없애기
plt.imshow(img_wordcloud) # 결과 보여주기
plt.savefig('동감_워드클라우드') # 파일 저장
```
