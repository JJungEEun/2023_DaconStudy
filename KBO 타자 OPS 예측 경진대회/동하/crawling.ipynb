```python
wd.get("http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=0&ys=2014&ye=2018&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=0&tr=&cv=&ml=1&sn=1000&si=&cn=")
#사용할 URL

html = wd.page_source
bsObject = BeautifulSoup(html, 'html.parser')
wd.quit()
```

```python
temp = bsObject.find_all("table")[1]
```

```python
column = ["순","이름","팀","WAR","G", "타석","타수","득점","안타","2타",
          "3타","홈런","루타","타점","도루","도실","볼넷","사구","고4", "삼진"
          ,"병살","희타","희비","타율","출루","장타","OPS","wOBA","WRC+"
          ]
df = pd.DataFrame(columns=column)
templen = len(temp.find_all("tr"))

for i in range(2, templen):
  tempTr = temp.find_all("tr")[i]
  if(tempTr.find("th")is not None):
    continue
  row = {}
  column_idx = 0
  for j in range(29):
    tempTd = tempTr.find_all("td")[j].text
    if(tempTd == "" and i >0):
      continue
    row[column[column_idx]] = tempTd
    column_idx += 1
  df = df.append(row, ignore_index = True)
```

```python
df
```

```python
df.drop("순", axis=1, inplace=True)
```

```python
df.to_csv("./data/14_18_hitter.csv", index=None)
```
